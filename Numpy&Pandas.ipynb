```python
"""
Use Case: Perfoorm the outlier detection for the given dataset i.e.,
datasetExample.csv
Dataset: datasetExample.csv
Perform the following task:
-> Load the data in the dataframe.
-> Detection of Outliers
"""
import pandas as pd
df = pd.read_csv("datasetExample.csv")
print("Dataset loaded successfully!\n")
print(df.head())
def detect_outliers_iqr(data):
    outlier_indices = {}
    for column in data.select_dtypes(include=['float64', 'int64']).columns:
        Q1 = data[column].quantile(0.25)
        Q3 = data[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
        outlier_indices[column] = outliers.index.tolist()
        print(f"\nColumn: {column}")
        print(f"Lower Bound = {lower_bound}, Upper Bound = {upper_bound}")
        print(f"Outliers detected: {outliers[column].values}")
    return outlier_indices
outlier_results = detect_outliers_iqr(df)
print("\nOutlier Indices (per column):")
print(outlier_results)
```
